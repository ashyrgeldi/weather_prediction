### 3. Training Machine Learning Model & Evaluation

```python
weather["target"] = weather.shift(-1)["temp_max"]
```

```python
weather
```

```python
weather = weather.iloc[:-1,:].copy()
```

```python
weather
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
predictors = ["precipitation", "temp_max", "temp_min"]
```

```python
train = weather.loc[:"2015-06-30"]
test = weather.loc["2015-07-01":]
```

```python
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(test["target"], predictions)
```

```python
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ["actual", "predictions"]
```
